# Neural Network / Deep Learning

```python
# Imports
# %matplotlib inline
import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow

from numpy.random import seed
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
```

```python
# Set the seed value for the notebook so the results are reproducible
seed(1)
```

```python
# Read the csv file into a pandas DataFrame
carbon = pd.read_csv('presolargrains_C_only_for_SVM.csv')
carbon.head()
```

## Data Preprocessing

```python
# Drop Type from the X values, use it as our dependent variable y
X = carbon.drop("Type", axis=1)
y = carbon["Type"]
print(X.shape, y.shape)
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
```

## Scale the data using MinMaxScalar

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Label encode the data set

```python
# Use LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
y_train_categorical
```

## Create a Deep Learning Model

```python
# Create model and add layers
model = Sequential()
```

```python
number_inputs = 1
number_hidden_nodes = 3
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
number_classes = 8
model.add(Dense(units=number_classes, activation='softmax'))
```

## Compile and train the model

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=300,
    shuffle=True,
    verbose=2
)
```

## Quantify the trained model

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Save the model

```python
# Save model to file
filename = 'deeplearning.sav'
joblib.dump(model, filename)
```
